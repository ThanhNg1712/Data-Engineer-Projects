```python
##Make a copy of the dataset
dataset_copy1 = dataset.copy()
dataset.columns
```

```python
##Drop column
col_to_drop = ['Start Date', 'End Date', 'Email Address','First Name', 'Last Name', 'Custom Data 1']
dataset_copy1 = dataset_copy1.drop(columns = col_to_drop)
print(dataset_copy1.shape)

```

## pandas.DataFrame.melt

```python
id_vars = list(dataset_copy1.columns)[0:8]
#id_vars
```

```python
value_vars = list(dataset_copy1.columns)[8:]
#value_vars
```

```python
dataset_melted = dataset_copy1.melt(id_vars =id_vars,value_vars = value_vars, var_name = "Question+ Sub-Question", value_name = "Answer")
dataset_melted
```

```python
path ="/content/gdrive/MyDrive/Colab Notebooks/My_Data/Data - Survey Monkey Output.xlsx"
question_import = pd.read_excel(path, sheet_name = "Question")
print(question_import.shape)
question_import.head()
```

```python
question = question_import.copy()
question.drop(columns=["Raw Question","Raw Subquestion", "Subquestion"], inplace=True)
question.rename(columns={"Question + Subquestion":"Question_Subquestion"},inplace=True )
question
```

```python
dataset_merged = pd.merge(left = dataset_melted, right = question, how = 'left', left_on ="Question+ Sub-Question",right_on ="Question_Subquestion")

print("melted", len(dataset_melted))
print("merge", len(dataset_merged))
print(dataset_melted.shape)
print(dataset_merged.shape)
dataset_merged
```

```python
#number of answer for each question
#eliminate Nan cell of answers where there is no reponse first
respondents = dataset_merged[dataset_merged['Answer'].notna()]
respondents=respondents.groupby("Question")["Respondent ID"].nunique().reset_index()
respondents.rename(columns = {"Respondent ID": "Respondents"}, inplace =True)
respondents
```

```python
dataset_merged_two = pd.merge(left = dataset_merged, right = respondents, how = 'left', left_on ="Question",right_on ="Question")

print("merge", len(dataset_merged))
print("merge 2", len(dataset_merged_two))
print(dataset_merged.shape)
print(dataset_merged_two.shape)
dataset_merged_two.head()
```

```python
# How many respondents answer the same reponse for each question
# So we need to count repondent id  group by question +subquestion

same_answers = dataset_merged #[dataset_merged['Answer'].notna()]
same_answers=same_answers.groupby(["Question_Subquestion", "Answer"])["Respondent ID"].nunique().reset_index()
same_answers.rename(columns = {"Respondent ID": "Same Answer"}, inplace =True)
same_answers
```

```python
#NAN value in the last column shows there is no response for that answer, replace with zero

dataset_merged_three = pd.merge(left = dataset_merged_two, right = same_answers, how = 'left', left_on =["Question_Subquestion", "Answer"],right_on =["Question_Subquestion", "Answer"])
dataset_merged_three['Same Answer'].fillna(0, inplace= True)
print("merge", len(dataset_merged_two))
print("merge 2", len(dataset_merged_three))
print(dataset_merged_two.shape)
print(dataset_merged_three.shape)
dataset_merged_three.head()
```

```python
output = dataset_merged_three.copy()
output.rename(columns = {"Identify which division you work in.-Response": "Division Primary",	"Identify which division you work in.-Other (please specify)":"Division Secodary",	"Which of the following best describes your position level?-Response": "Position",	
 "Which generation are you apart of?-Response": "Generation",	"Please select the gender in which you identify.-Response" : "Gender",	"Which duration range best aligns with your tenure at your company?-Response":"Tenure",	
 "Which of the following best describes your employment type?-Response": "Employment Type"}, inplace =True)
output
```
